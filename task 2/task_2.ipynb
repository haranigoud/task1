```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")

print("First 5 Rows")
display(df.head())

print("Last 5 Rows")
display(df.tail())

numerical_cols = ['math score', 'reading score', 'writing score']
categorical_cols = [
    'gender',
    'race/ethnicity',
    'parental level of education',
    'lunch',
    'test preparation course'
]
ordinal_cols = ['parental level of education']
binary_cols = ['gender', 'lunch', 'test preparation course']

print("Numerical Columns:")
display(df[numerical_cols].head())

print("Categorical Columns:")
display(df[categorical_cols].head())

print("Ordinal Columns:")
display(df[ordinal_cols].tail())

print("Binary Columns:")
display(df[binary_cols].head())

info_table = pd.DataFrame({
    "Column": df.columns,
    "Non-Null Count": df.notnull().sum().values,
    "Data Type": df.dtypes.values
})
print("Dataset Info")
display(info_table)

print("Dataset Description")
display(df.describe(include="all"))

unique_summary = pd.DataFrame({
    "Column": categorical_cols,
    "Unique Values Count": [
        df[col].nunique(dropna=False) for col in categorical_cols
    ]
})
print("\nSummary of Unique Values Count:")
display(unique_summary)

target_variable = "math score"
print("\nTarget Variable:", target_variable)

input_features = df.drop(columns=[target_variable]).columns.tolist()
print("Input Features:", input_features)

num_rows, num_cols = df.shape
print("Dataset Size:")
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_cols}")

print("Missing Values")
display(df.isnull().sum().to_frame("Missing Count"))

print("Math Score Distribution")
display(df["math score"].describe().to_frame("Statistics"))
```
